```python
# Import Dependencies 

import os
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import folium
from shapely.geometry import Point
```

Load the data 

```python
# Load files 
file_path = "C:\\Users\\linet\\OneDrive\\Desktop\\olympics\\Olympics-\\athlete_events.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.shape)
df.head()
```

```python
df.columns
```

Clean the data 

```python
# Drop the specified columns
df = df.drop(columns=['Height', 'Weight', 'Year'])
df.head()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Replace missing values with 0 in the specified columns
# Replace missing values in object columns
df[['Name', 'Sport', 'Medal']] = df[['Name', 'Sport', 'Medal']].fillna('Unknown')
df['Sex'] = df['Sex'].fillna('O')  # Replace NaN in Sex with 'O'
df['Age'] = pd.to_numeric(df['Age'], errors='coerce').fillna(0)
df.head()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Display summary statistics for numerical columns
df.describe()
```

```python
# Display unique values in the 'Sport' column
df['Sport'].unique()
```

```python
# Display unique values in the 'NOC' column
unique_nocs = df['NOC'].unique()
unique_nocs
```

```python
# Create DataFrame from NOC codes
noc_df = pd.DataFrame(unique_nocs, columns=['NOC'])
noc_df.head()
```

```python
# Display unique values and their counts in the 'Medal' column
medal_counts = df['Medal'].value_counts()
medal_counts
```

```python
# Count the number of unique athletes by medal type
medal_unique_counts = df.groupby('Medal')['ID'].nunique()

# Drop the 'Unknown' medal type if it exists
medal_unique_counts = medal_unique_counts.drop('Unknown', errors='ignore')
medal_unique_counts 
```

```python
# Plot the counts of each medal type
plt.figure(figsize=(10, 6))
sns.barplot(x=medal_unique_counts.index, y=medal_unique_counts.values, palette="magma")
plt.title('Unique Athletes per Medal Type')
plt.xlabel('Medal Type')
plt.ylabel('Number of Unique Athletes')
plt.show()
```

```python
# Group by Medal and Team, and count unique athletes
medal_country_unique_counts = df.groupby(['Medal', 'Team'])['ID'].nunique().reset_index()

# Pivot the table for better plotting
pivot_table = medal_country_unique_counts.pivot(index='Team', columns='Medal', values='ID').fillna(0)

# Add a total column to the pivot table
pivot_table['Total'] = pivot_table.sum(axis=1)

# Sort the table by the total column in descending order
sorted_pivot_table = pivot_table.sort_values(by='Total', ascending=False)

# Select the top 100 countries
top_100_countries = sorted_pivot_table.head(100)
top_100_countries.head()
```

```python
# Number of subplots needed
num_plots = 10
countries_per_plot = 10

# Create subplots
fig, axes = plt.subplots(num_plots, 1, figsize=(20, 50), sharex=True)

for i in range(num_plots):
    start_idx = i * countries_per_plot
    end_idx = start_idx + countries_per_plot
    subset = top_100_countries.iloc[start_idx:end_idx].drop(columns='Total')  # Drop 'Total' for this subset

    subset.plot(kind='bar', stacked=True, ax=axes[i], colormap='viridis')
    axes[i].set_title(f'Countries {start_idx + 1} to {end_idx}')
    axes[i].set_ylabel('Number of Unique Athletes')
    axes[i].legend(title='Medal', loc='upper right')

# Set common labels and titles
axes[-1].set_xlabel('Country')
fig.suptitle('Top 100 Countries with Medals', fontsize=16)
plt.xticks(rotation=90)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

```python
# Plot the distribution of athletes' ages
plt.hist(df['Age'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Athletes\' Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group by Sport, Sex, and Age, and count unique athletes
participation_by_sport_sex_age = df.groupby(['Sport', 'Sex', 'Age'])['ID'].nunique().reset_index()

# Pivot the table for better plotting
pivot_table_sex = participation_by_sport_sex_age.pivot_table(index='Sport', columns='Sex', values='ID', aggfunc='sum').fillna(0)
pivot_table_age = participation_by_sport_sex_age.pivot_table(index=['Sport', 'Sex'], columns='Age', values='ID', aggfunc='sum').fillna(0)

# Plotting heatmap for participation by sport and sex
plt.figure(figsize=(14, 10))
sns.heatmap(pivot_table_sex, cmap='YlGnBu', annot=True, fmt='g', linewidths=0.5, cbar_kws={'label': 'Number of Athletes'})
plt.title('Number of Athletes by Sport and Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Sport', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
        
```

```python
# Plotting bar plots for participation by age for each sport and sex
sports = df['Sport'].unique()
num_sports = len(sports)
fig, axes = plt.subplots(num_sports, 1, figsize=(14, num_sports * 3), sharex=True)

for i, sport in enumerate(sports):
    subset = participation_by_sport_sex_age[participation_by_sport_sex_age['Sport'] == sport]
    sns.barplot(data=subset, x='Age', y='ID', hue='Sex', ax=axes[i], palette='magma')
    axes[i].set_title(f'Participation by Age in {sport}', fontsize=14)
    axes[i].set_ylabel('Number of Athletes')
    axes[i].legend(title='Gender')

plt.xlabel('Age', fontsize=14)
plt.tight_layout()
plt.show()
```
